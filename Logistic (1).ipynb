```python
df = pd.read_csv("sgemm_product.csv")
#df.head()
df["RunF"]= (df["Run1 (ms)"] + df["Run2 (ms)"]+ df["Run3 (ms)"]+df["Run4 (ms)"])/4
df=df.drop(columns=['Run1 (ms)', 'Run2 (ms)', 'Run3 (ms)', 'Run4 (ms)'])
#df.head()
dfl=df
df['RunF'].mean()
dfl['RunF_binary'] = np.where(df['RunF']>=3.7878695764981335e-15, 1, 0)
#logistic Dataset
df_log=dfl.drop(columns=['RunF'])
#replicate test and train for logistic regression with binary y 
from sklearn.model_selection import train_test_split 
trainl,testl = train_test_split(df_log,test_size = 0.4, random_state=5000)
```

```python


X_lr = trainl.iloc[:, 0:14]
oneslr = np.ones([X_lr.shape[0],1])
X_lr = np.concatenate((oneslr,X_lr),axis=1)

y_lr =  trainl.iloc[:, -1].values #.values converts it from pandas.core.frame.DataFrame to numpy.ndarray
theta_lr = np.zeros([1,15])

#set hyper parameters
alpha = 0.5
iters = 5
```

```python

#define sigmoid for predicted y 
def sigmoid(z):
    return 1 / (1 + np.exp(-z))


def predicted_y(X_lr, theta_lr):
    return sigmoid(np.dot(X_lr, theta_lr.T))

#computecost logistic
def computecost_l(h, y_lr):
    return (-y_lr * np.log(h) - (1 - y_lr) * np.log(1 - h)).mean()



```
